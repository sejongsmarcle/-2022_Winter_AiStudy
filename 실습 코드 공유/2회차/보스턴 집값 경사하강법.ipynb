# 라이브러리 import

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
from sklearn.datasets import load_boston
%matplotlib inline
```

# 데이터 로드

```python
bst = load_boston()
bstDF = pd.DataFrame(bst.data, columns = bst.feature_names)
# shape는 머신러닝에서 행렬의 차원을 표현한 것이다. (numpy 제공)
print(bstDF.shape)
# (506, 13) 은 506개의 행과 13의 열을 말하는 것이다.
```

```python
bstDF.head() # 데이터 확인 (pandas 제공)
```

# 독립변수로 사용할 변수 선택

```python
bstDF1 = bstDF[['RM', 'PTRATIO']] # 거주할 수 있는 방 개수, 지역의 교사와 학생 수 비율
print(bstDF1.shape)
bstDF1.head()
```

```python
bstDF2 = bstDF[['CRIM', 'NOX']] # 지역별 범죄 발생률, 일산화질소 농도
print(bstDF2.shape)
bstDF2.head()
```

```python
# to_numpy() 메서드, pandas 객체 -> numpy 객체인 ndarray로 변환한다.
x_data1 = bstDF1.to_numpy()
x_data2 = bstDF2.to_numpy()
# target => 종속변수 뽑을 때 사용
y_data = bst.target

print(x_data1[:5][:])
print(x_data2[:5][:])
print(y_data[:5])
```

# 기울기, 편향 값 초기화

```python
# zeros는 0으로 가득한 array를 말함
# np.zeros(shape, dtype, order) 형태
slopes1 = np.zeros(x_data1.shape[1], dtype=np.float32) # a_1, a_2 기울기 해당
b1 = 0. # 편향

slopes2 = np.zeros(x_data2.shape[1], dtype=np.float32) # a_1, a_2 기울기 해당
b2 = 0. # 편향

lr = 0.0001 # 학습률

epochs = 5001

print("기울기1: {}, 편향1: {}".format(slopes1, b1))
print("기울기2: {}, 편향2: {}".format(slopes2, b2))
```

## 변수 값의 범위를 조정

```python
# 사이킷런: 머신러닝 분석할 때 유용한 라이브러리
# MinMaxScaler는 스케일을 조정하는 정규화 함수, 모든 데이터가 0과 1 사이의 값을 갖도록 해주는 함수.
# -> 따라서 최댓값는 1로, 최솟값은 0으로 데이터의 범위를 조정해준다.
from sklearn.preprocessing import MinMaxScaler

# fit: 훈련 데이터에만 적용해, 훈련 데이터의 분포를 먼저 학습
# transform: 훈련 데이터와 테스트 데이터에 적용해 스케일 조정
 
scaler1 = MinMaxScaler()
scaler1.fit(x_data1)
x_data1 = scaler1.transform(x_data1)

scaler2 = MinMaxScaler()
scaler2.fit(x_data2)
x_data2 = scaler2.transform(x_data2)

print("x_data1:\n", x_data1)
print("x_data2:\n", x_data2)
```

## 경사 하강법으로 모델 학습시키기

```python
for i in range(epochs):
  y_pred = x_data1.dot(slopes1) + b1 # y = a1 * x1 + a2 * x2 + a3 * x3 + b와 같다.
  # dot은 numpy array를 곱할 때 사용하고, 행렬의 곱으로 생각하면 된다.
  error = y_data - y_pred
    
  a_diff = np.array([-(2 / len(x_data1)) * sum(x_data1[:, i] * (error)) for i in range(len(slopes1))])
  b_diff = -(2 / len(x_data1) ) * sum(error)
    
  slopes1 -= lr * a_diff
  b1 -= lr * b_diff
   
  if i % 100 == 0:
      print("epoch{:04d}, 기 울 기: {}, 절 편: {:04f}".format(i, slopes1, b1))
      print("a1, a2의 편미분: {}, b의 편미분: {} \n".format(a_diff, b_diff))
```

```python
for i in range(epochs):
    y_pred = x_data2.dot(slopes2) + b2
    error = y_data - y_pred
    
    a_diff = np.array([-(2 / len(x_data2)) * sum(x_data2[:, i] * (error)) for i in range(len(slopes2))])
    b_diff = -(2 / len(x_data2) ) * sum(error)
    
    slopes2 -= lr * a_diff
    b2 -= lr * b_diff
    
    if i % 100 == 0:
        print("epoch{:04d}, 기 울 기: {}, 절 편: {:04f}".format(i, slopes2, b2))
        print("a1, a2의 편미분: {}, b의 편미분: {} \n".format(a_diff, b_diff))
```

## 예측한 집 값

```python
y_pred1 = x_data1.dot(slopes1) + b1

y_pred2 = x_data2.dot(slopes2) + b2

print("y_pred1:\n", y_pred1)
print("y_pred2:\n", y_pred2)
```

## 예측한 집 값 시각화
